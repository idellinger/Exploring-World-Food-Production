# Bananas
Banana production has been closeley tied with political and economic developments in South America over the last 150 years. 

```python
from matplotlib.ticker import FuncFormatter

fig, ax = plt.subplots(ncols=3,nrows=2,figsize=(15,8))

countries_banana = {'Afghanistan','Ecuador','United States','France','Dominican Republic','China'}

# Function to format y-axis labels in 'Tonnes'
def tonnage_formatter(value, _):
    return f'{value/1000:.0f} Tonnes'

for i, country in enumerate(countries_banana):
    row = i // 3  # Calculate the row index
    col = i % 3   # Calculate the column index

    # Filter data for the specific country
    country_data = df[df['Country'] == country]

    
    ax[row, col].plot(country_data['Year'], country_data['Bananas'], label=country)
    ax[row, col].set_title(country)
    ax[row, col].set_xlabel('Year')
    ax[row, col].set_ylabel('Banana Production (Tonnes)')

    ax[row, col].yaxis.set_major_formatter(FuncFormatter(tonnage_formatter))

plt.tight_layout()

plt.show()



```

Ecuador is one of the biggest exporters of Bananas so something is off with the data, lets dive deeper.

```python
# Filter the DataFrame for banana data from Ecuador
ecuador_bananas = df.loc[df['Country'] == 'Ecuador', ['Year', 'Bananas']]
# Sort by 'Year' numerically
ecuador_bananas_sorted = ecuador_bananas.sort_values(by='Year')

# Display the sorted DataFrame
print(ecuador_bananas_sorted)
```

```python

# Filter the DataFrame for the year 1964 and exclude countries in the ignore list
banana_1964 = df[(df['Year'] == 1964) & (~df['Country'].isin(country_rows_ignore))]

# Calculate percentages
banana_1964['Percentage'] = banana_1964['Bananas'] / banana_1964['Bananas'].sum() * 100

# Group countries with less than 1% into 'Other, less than 1%'
threshold = 1
banana_1964.loc[banana_1964['Percentage'] < threshold, 'Country'] = 'Other, less than 1%'
banana_1964 = banana_1964.groupby('Country')['Bananas'].sum().reset_index()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(banana_1964['Bananas'], labels=banana_1964['Country'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title('Total Banana Production by Country in 1964 (Grouping < 1%)')
plt.show()
```

```python
# Filter the DataFrame for the year 2000 and exclude countries in the ignore list
banana_2000 = df[(df['Year'] == 2000) & (~df['Country'].isin(country_rows_ignore))]

# Calculate percentages
banana_2000['Percentage'] = banana_2000['Bananas'] / banana_2000['Bananas'].sum() * 100

# Group countries with less than 1% into 'Other, less than 1%'
threshold = 1
banana_2000.loc[banana_2000['Percentage'] < threshold, 'Country'] = 'Other, less than 1%'
banana_2000 = banana_2000.groupby('Country')['Bananas'].sum().reset_index()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(banana_2000['Bananas'], labels=banana_2000['Country'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title('Total Banana Production by Country in 2000 (Grouping < 1%)')
plt.show()
```

In both of the above charts, Ecuador's banana production doesn't display. Yet, in the years 1964 and 2000, Ecuador exported ~25% and ~28%, respectively, of all the Bananas traded worldwide (https://www.hrw.org/reports/2002/ecuador/ecuad0402-03.htm).

According to data provided by CONABAN, Ecuador exported 4,543,556 metric tons of bananas from January through November 2000. CONABAN-Ecuador, "La Industria Bananera: . . . ," p. 12
